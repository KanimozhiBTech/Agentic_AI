## Model integration using Google Gemini, Groq 

```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ["OPENAI_API_KEY"]=os.getenv("OPENAI_API_KEY")
os.environ["GOOGLE_API_KEY"]=os.getenv("GOOGLE_API_KEY")
os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
```

```python
from langchain.chat_models import init_chat_model

model=init_chat_model("google_genai:gemini-2.5-flash-lite")
response=model.invoke("Tell me about yourself")
print(response)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI

model=ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
response=model.invoke("what is machine learning?")
print(response.content)
```

### GROQ Model Integration

```python
from langchain.chat_models import init_chat_model

model=init_chat_model(model="groq:qwen/qwen3-32b")
response=model.invoke("essay about health")
print(response.content)
```

```python
from langchain_groq import ChatGroq

model=ChatGroq(model="qwen/qwen3-32b")
response=model.invoke("write a poem on unity")
print(response.content)
```

### Streaming

```python
model.stream("Write me a 200 words paragraph on Artificial Intelligence")
```

```python
for chunk in model.stream("Write me a 200 words paragraph on Artificial Intelligence"):
    print(chunk.text,end="|",flush=True)
```

### Batching

```python
responses=model.batch([
    "write a essay about national bird of india",
    "write a essay about national animal of india",
    "write a essay about national tree of india"
])

```

```python
responses
```

```python
for response in responses:
    print(response)
```

```python
responses=model.batch([
    "write a essay about national bird of india",
    "write a essay about national animal of india",
    "write a essay about national tree of india"
],
config={"max_concurrency":5})

for response in responses:
    print(response)
```
