```python
from langchain.chat_models import init_chat_model
# from dotenv import load_dotenv
# import os

# load_dotenv()

# os.environ["GROQ_API_KEY"]=os.getenv("GROQ_API_KEY")
model=init_chat_model(model="groq:qwen/qwen3-32b")
response=model.invoke("what is linear algebra?")
response
```

```python
from langchain.tools import tool

@tool
def get_weather(city:str)->str:
    """get weather of a city"""
    return f"the weather of {city} is sunny"

model_with_tool=model.bind_tools([get_weather])
response=model_with_tool.invoke("what is the weathear in boston?")
response
```

```python
for tool in response.tool_calls:
    print(f"Name {tool['name']}")
    print(f"Args {tool['args']}")
    print(f"Id {tool['id']}")
    print(f"Type {tool['type']}")
```

```python
messages=[{"role":"user","content":"What's the weather in bengaluru?"}]
ai_msg=model_with_tool.invoke(messages)
messages.append(ai_msg)

for tool_call in ai_msg.tool_calls:
    tool_result=get_weather.invoke(tool_call)
    messages.append(tool_result)

final_result=model_with_tool.invoke(messages)
print(final_result.text)
```
